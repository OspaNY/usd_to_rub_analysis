```python
import pandas as pd
```

Коды валют и котировки на данный момент

```python
start = '01/01/2017'
stop = '15/07/2025'
url1 = 'http://www.cbr.ru/scripts/XML_daily.asp?date_req=' + stop
pd.read_xml(url1, encoding='cp1251')
```

Курс доллара

```python
url = 'https://www.cbr.ru/scripts/XML_dynamic.asp?date_req1=' + start + '&date_req2=' + stop + '&VAL_NM_RQ=R01235'
df = pd.read_xml(url)
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)  # дата в формате дд.мм.гггг
df['usd_rub'] = df['VunitRate'].str.replace(',', '.').astype(float)

df.drop(["Id", "Nominal", "Value", 'VunitRate'], axis=1, inplace = True)
print(df)
```

```python
df.to_csv('data/usd_rub.csv', index = False)  
```
